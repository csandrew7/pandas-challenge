```python
# Using groupby to group into gender
# Count purchases by gender
gender_pa = purchase_data.groupby(["Gender"])
gender_pcount = gender_pa["Gender"].count()
print(gender_pcount)
print("----------")
# Calculate average purchase price by gender
gender_avg = gender_pa["Price"].mean()
print(gender_avg)
print("----------")
# Calculate total purchase value
gender_total_value = gender_pa["Price"].sum()
print(gender_total_value)
print("----------")
# Calculate average total purchase per person
avg_total_person = gender_total_value / gender.groupby(["Gender"]).count()["Price"]
print(avg_total_person)
```

```python
pd.DataFrame({
    "Purchase Count" : gender_pcount,
    "Average Purchase Price" : gender_avg.map('${:,.2f}'.format),
    "Total Purchase Value" : gender_total_value.map('${:,.2f}'.format),
    "Avg Total Purchase per Person" : avg_total_person.map('${:,.2f}'.format)
}, index=purchase_data.Gender.unique()
).sort_index()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data.head()
```

```python
purchase_data["Age"].describe()
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
age = pd.cut(purchase_data["Age"], bins, labels = group_labels)
age
```

```python
purchase_data["Age_Group"] = age
purchase_data.head()
```

```python
players_age = purchase_data.groupby("Age_Group").SN.nunique()
print(players_age)
print("----------")
pct_pa = (players_age / purchase_data.SN.nunique()) * 100
print(pct_pa)
```

```python
pd.DataFrame({
    "Total Count" : players_age,
    "Percentage of Players" : pct_pa.map('{:,.2f}%'.format),
}, index= group_labels
)
```

```python
sn_total = purchase_data["SN"].value_counts()
print(sn_total)
print("----------")
spenders_total = purchase_data.groupby("SN")
spenders_pcount = spenders_total["Purchase ID"].count()
print(spenders_pcount)
print("----------")
spenders_avg = round(spenders_total["Price"].mean(),2)
print(spenders_avg)
print("----------")
spenders_total = spenders_total["Price"].sum()
print(spenders_total.head())
print("----------")
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_list = purchase_data[["Item ID", "Item Name", "Price"]]

popular_items = popular_items_list.groupby(["Item ID", "Item Name"])

ip_pcount = popular_items["Price"].count()

ip_price = popular_items["Price"].sum()

ip_value = ip_price / ip_pcount

most_popular_items = pd.DataFrame({
    "Purchase Count" : ip_pcount,
    "Item Price" : ip_value,
    "Total Purchase Value" : ip_price
})
most_popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
sorted_items = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()
sorted_items["Item Price"] = sorted_items["Item Price"].astype(float).map("${:,.2f}".format)
sorted_items["Total Purchase Value"] = sorted_items["Total Purchase Value"].astype(float).map("${:,.2f}".format)
sorted_items.head()
```
